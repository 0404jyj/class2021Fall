### Tokenization

어떤 문장이 있을 때 그것을 단어를 기준으로 쪼개 주는 것임.

```python
import nltk
```

```python
text = 'Here’s to the crazy ones, the misfits, the rebels, the troublemakers, the round pegs in the square holes. \
The ones who see things differently — they’re not fond of rules. \
You can quote them, disagree with them, glorify or vilify them, \
but the only thing you can’t do is ignore them because they change things. \
They push the human race forward, and while some may see them as the crazy ones, we see genius, \
because the ones who are crazy enough to think that they can change the world, are the ones who do.'
```

```python
text.split()
```

```python
' '.join(text.split())
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')  # 펑션을 단어가 아닌걸 체크할 수 있게 해줌
words = word_tokenize(text)
words
```

```python
from nltk.tokenize import RegexpTokenizer  # Regular expression의 규칙을 지켜서 토크나이즈 할 수 있게 할 수 잇음
retokenize = RegexpTokenizer("[\w]+") #여기서 규칙을 설정 - 단어들만 나오게 해줌
words = retokenize.tokenize(text)
```

### Normalization  
Stemming: am → am, the going → the go, having → hav  그냥 뒷부분을 끝어주는 것 
Lemmatization: am → be, the going → the going, having → have  진짜 원형으로 만들어 주는 것

변이형들을 하나의 형태로 즉 어근으로 추린다.

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
[stemmer.stem(w) for w in words]  #  그냥 뒷부분 짤라주는 것
```

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words]
```

```python
from nltk.stem import WordNetLemmatizer # 원형으로 만들어 주는 것
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(w) for w in words]
```

### Stopword

빈번한게 나오는 단어들은 카운트 하지 않겠다.

```python
from nltk.corpus import stopwords  
nltk.download('stopwords')
print(words)
words = [w for w in words if not w in stopwords.words('english')]  # stopwords들이 아니라면 print 하라 
print(words)
```

### Collocation, Concordance

어떤 두 단어의 연쇄를 보는 것

예문을 찾아볼 때 유용한 것...?



```python
nltk.download('gutenberg')
text = nltk.corpus.gutenberg.raw('austen-emma.txt')
words = retokenize.tokenize(text) # 문장을 단어로 나누는 과정
```

```python
nltk.Text(words).collocations() #  빈번하게 나오는 두 단어를 체크
```

```python
nltk.Text(words).concordance('should', 79, 10) # 예문을 찾는다. (찾는 단어, 단어 포함 문자열 수, 용례 수)
```

```python
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])

#이 소설 속에 어떠한 단어가 어떻게 분포 되어 있는가?
```

```python
nltk.Text(words).similar("Emma") # 특정 단어와 비슷하게 쓰이는 것을 찾기
```

```python
nltk.Text(words).common_contexts(["Emma", "she"])  #두 단어가 비슷하게 나오는 경우 
```

```python
import re
```

```python
word = 'supercalifragilisticepialidocious'
re.search('[aeu]li', word) # []li가 들어 있는 것을 찾아라 ex) a li e li u li
```

```python
re.findall('{ai}', 'considered condo')
```

```python
engdict = nltk.corpus.words.words('en')

result1 = [w for w in engdict if re.search('ed$', w)]
result2 = [w for w in engdict if re.search('^..j..t..$', w)] # 아무거나 두개 j 아무거나 두개 t 아무거나 두개 
result3 = [w for w in engdict if re.search('^[ghi][mno][jlk][def]$', w)]
result4 = [w for w in engdict if re.search('^[ah]+$', w)]
print(result1[:10])
print(result2[:10])
print(result3[:10])
print(result4[:10])
```

### Extract information (pos tag, named entity)

품사를 체크 해주는 것. 카테고리를 이야기 해주는 것.

```python
sent = "I am Jhon from America and would like to go to Starbucks"
words = nltk.word_tokenize(sent)
```

```python
'''
POS tag list:

CC	coordinating conjunction
CD	cardinal digit
DT	determiner
EX	existential there (like: "there is" ... think of it like "there exists")
FW	foreign word
IN	preposition/subordinating conjunction
JJ	adjective	'big'
JJR	adjective, comparative	'bigger'
JJS	adjective, superlative	'biggest'
LS	list marker	1)
MD	modal	could, will
NN	noun, singular 'desk'
NNS	noun plural	'desks'
NNP	proper noun, singular	'Harrison'
NNPS	proper noun, plural	'Americans'
PDT	predeterminer	'all the kids'
POS	possessive ending	parent's
PRP	personal pronoun	I, he, she
PRP$	possessive pronoun	my, his, hers
RB	adverb	very, silently,
RBR	adverb, comparative	better
RBS	adverb, superlative	best
RP	particle	give up
TO	to	go 'to' the store.
UH	interjection	errrrrrrrm
VB	verb, base form	take
VBD	verb, past tense	took
VBG	verb, gerund/present participle	taking
VBN	verb, past participle	taken
VBP	verb, sing. present, non-3d	take
VBZ	verb, 3rd person sing. present	takes
WDT	wh-determiner	which
WP	wh-pronoun	who, what
WP$	possessive wh-pronoun	whose
WRB	wh-abverb	where, when
'''
```

```python
nltk.download('averaged_perceptron_tagger')
pos = nltk.pos_tag(words)
pos
```

```python
nltk.download('maxent_ne_chunker')
NE = nltk.ne_chunk(pos)
# common Entity types: ORGANIZATION, PERSON, LOCATION, DATE, TIME, MONEY, and GPE (geo-political entity)
```

```python
print(NE)
```

### Wordcloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = nltk.corpus.gutenberg.raw('bible-kjv.txt')

wc = WordCloud().generate(text) 
plt.imshow(wc)
```

```python
stopwords = set(STOPWORDS) 
stopwords.add('unto') # 추가로 stopwords에 추가하는 것
wc = WordCloud(stopwords = stopwords).generate(text) # stopwords에 포함된 문자 빼주는 것.
plt.imshow(wc)
```
